# Data - cpu_bounded_func

```python
import pickle
data = pickle.load(open('cpu_bounded_func.pickle', 'rb'))
```

```python
for i, sim in enumerate(data):
    data[i] = eval(sim)
```

```python
lambdas = [
    'py_lambda_concurrency_stress_768',
    'py_lambda_concurrency_stress_1280',
    'py_lambda_concurrency_stress_1792',
    'py_lambda_concurrency_stress_2304',
    'py_lambda_concurrency_stress_2816'
]

lambdas = [*lambdas, *lambdas, *lambdas]
```

```python
for i, sim in enumerate(data):
    data[i]['lambda'] = lambdas[i]
```

```python
import pandas as pd

tab_data = []

columns = [
    'lambda',
    'executor',
    'final_time',
    'memory',
    'cpu_count',
    'pool_size',
    'logical_cpu_count'
]

for i, sim in enumerate(data):
    row = []
    row.append(sim['lambda'])
    row.append(sim['executor'])
    row.append(float(sim['final_time']))
    row.append(int(sim['lambda'].split('_')[4]))
    row.append(int(sim['cpu_count']))
    row.append(int(sim['pool_size']))
    row.append(int(sim['logical_cpu_count']))
    tab_data.append(row)
    
tab = pd.DataFrame(tab_data, columns=columns)
```

```python
import numpy as np

tab['billed_time'] = np.ceil(tab['final_time'] * 1000.0)
tab['executor'] = tab['executor'].astype('category')
tab['memory'] = tab['memory'].astype('category')

tab
```

```python
import plotly.express as px
fig = px.bar(tab, x='executor', y='billed_time', color='memory', barmode='group')
fig.update_layout(title_text='cpu_bounded_func')
fig.write_image("cpu_bounded_func_billed_time.png")
```
