# Data - cpu_bounded_func

```python
import pickle
data = pickle.load(open('io_bounded_func_200.pickle', 'rb'))
```

```python
lambdas = [
    'py_lambda_concurrency_stress_768',
    'py_lambda_concurrency_stress_1280',
    'py_lambda_concurrency_stress_1792',
    'py_lambda_concurrency_stress_2304',
    'py_lambda_concurrency_stress_2816'
]

lambdas = [*lambdas, *lambdas, *lambdas]
```

```python
for i, sim in enumerate(data):
    data[i]['lambda'] = lambdas[i]
```

```python
import pandas as pd

tab_data = []

columns = [
    'lambda',
    'executor',
    'final_time',
    'memory',
    'cpu_count',
    'pool_size',
    'tot_number',
    'logical_cpu_count'
]

for i, sim in enumerate(data):
    row = []
    row.append(sim['lambda'])
    row.append(sim['executor'])
    row.append(float(sim['final_time']))
    row.append(int(sim['lambda'].split('_')[4]))
    row.append(int(sim['cpu_count']))
    row.append(int(sim['pool_size']))
    row.append(int(sim['tot_number']))
    row.append(int(sim['logical_cpu_count']))
    tab_data.append(row)
    
tab = pd.DataFrame(tab_data, columns=columns)
```

```python
import numpy as np

tab['billed_time'] = np.ceil(tab['final_time'] * 1000.0)
tab['executor'] = tab['executor'].astype('category')
tab['memory'] = tab['memory'].astype('category')

tab
```

```python
import plotly.express as px
fig = px.bar(tab, x='executor', y='billed_time', color='memory', barmode='group')
fig.update_layout(title_text='io_bounded_func x 200')
fig.write_image("io_bounded_func_billed_time.png")
```

```python
data_2000 = pickle.load(open('io_bounded_func_2000.pickle', 'rb'))
```

```python
tab_data_2000 = []

columns = [
    'lambda',
    'executor',
    'final_time',
    'memory',
    'cpu_count',
    'pool_size',
    'tot_number',
    'logical_cpu_count'
]

for i, sim in enumerate(data_2000):
    row = []
    row.append(sim['lambda'])
    row.append(sim['executor'])
    row.append(float(sim['final_time']))
    row.append(int(sim['lambda'].split('_')[4]))
    row.append(int(sim['cpu_count']))
    row.append(int(sim['pool_size']))
    row.append(int(sim['tot_number']))
    row.append(int(sim['logical_cpu_count']))
    tab_data_2000.append(row)

tab_data_2000 = pd.DataFrame(tab_data_2000, columns=columns)
```

```python
tab_data_2000['billed_time'] = np.ceil(tab_data_2000['final_time'] * 1000.0)

data_2000 = pd.concat([tab_data_2000, tab[tab['lambda']=="py_lambda_concurrency_stress_1792"]])

data_2000['executor'] = data_2000['executor'].astype('category')
data_2000['memory'] = data_2000['memory'].astype('category')
data_2000['tot_number'] = data_2000['tot_number'].astype('category')

data_2000 = data_2000.sort_values(by=['tot_number'])
data_2000
```

```python
import plotly.express as px
fig = px.bar(data_2000, x='executor', y='billed_time', color='tot_number', barmode='group')
fig.update_layout(title_text='io_bounded_func')
fig.write_image("io_bounded_func_tot_number.png")
```
